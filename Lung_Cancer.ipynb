Categorical features like Smoking_History and Tumor_Location need to be converted into a numeric format before they can be used in regression models. However, using a LabelEncoder for these features is generally not recommended because it introduces an artificial ordinal relationship between the encoded values. Instead, it's better to use one-hot encoding for nominal categorical variables to avoid imposing any such order. Here's how you can preprocess these features:

Data Preprocessing Steps
One-Hot Encoding for Nominal Categorical Variables:

For categorical features without any inherent order (like Smoking_History and Tumor_Location), use one-hot encoding.
Ordinal Encoding:

For ordinal features (if any), you can use LabelEncoder or a custom ordinal encoding. For example, if Smoking_History has an inherent order (e.g., Never, Former, Current), you might assign Never = 0, Former = 1, and Current = 2.
Standardization/Normalization:

Standardize or normalize the continuous features to ensure they have a similar scale.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

When Normalization Might Not Be Necessary
Tree-Based Algorithms: Algorithms like decision trees, random forests, and gradient boosting trees do not require normalization as they are not sensitive to the scale of the features.
Linear Models: For logistic regression, while normalization can improve training speed and stability, the model can still perform well without it, especially if the features are already on a similar scale or if the dataset size is large enough.

```python
# Normalising the values inside Tumor_Size_mm
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
scaler  = MinMaxScaler()
df['Normalised_Tumor_Size'] = scaler.fit_transform(df[['Tumor_Size_mm']])
df['Tumor_Size_mm' ]= df['Normalised_Tumor_Size']
%matplotlib inline
plt.bar(df['Age'][:10],df['Tumor_Size_mm'][:10], color='skyblue')
plt.xlabel('Age')
plt.ylabel('Tumor Size')
# plt.show()
df.drop(columns=['Normalised_Tumor_Size'])
```

```python
# Normalising  Alanine_Aminotransferase_Level
df['Normalised_Alanine_Levels'] = scaler.fit_transform(df[['Alanine_Aminotransferase_Level']])
df['Alanine_Aminotransferase_Level'] = df['Normalised_Alanine_Levels'] 
```

```python
del df['Normalised_Tumor_Size']
```

```python
df.head(5)
```

```python
# Normalising all other factors using loop 
from sklearn.preprocessing import MinMaxScaler

# Columns to be normalized
columns_to_normalize = [
    'Aspartate_Aminotransferase_Level',
    'Creatinine_Level',
    'LDH_Level',
    'Calcium_Level',
    'Phosphorus_Level',
    'Glucose_Level',
    'Potassium_Level',
    'Sodium_Level'
]

# Initialize the scaler
scaler = MinMaxScaler()

# Loop through each column and normalize
for column in columns_to_normalize:
    # Create a new column with normalized values
    df[f'Normalized_{column}_Level'] = scaler.fit_transform(df[[column]])
    # Replace the original column with normalized values
    df[column] = df[f'Normalized_{column}_Level']

# Optionally, you can drop the intermediate columns if you don't need them anymore
for column in columns_to_normalize:
    df.drop(columns=[f'Normalized_{column}_Level'], inplace=True)
```

```python
#Let Us Normalise Survival_Months and Smocking_Pack_years

columns_to_normalize_2 = [
    'Survival_Months',
    'Smoking_Pack_Years'
]

for it in columns_to_normalize_2:
    df[f'Normalised_{it}'] = scaler.fit_transform(df[[f'{it}']])
    df[it]  = df[f'Normalised_{it}']
    del df[f'Normalised_{it}']
```

```python
x_columns = [
    'Tumor_Size_mm',
    'Survival_Months',
     'Alanine_Aminotransferase_Level',
     'Aspartate_Aminotransferase_Level',
     'Creatinine_Level',
     'LDH_Level',
      'Calcium_Level',
      'Phosphorus_Level',
    'Glucose_Level',
     'Potassium_Level',
     'Sodium_Level',
]
df2 = pd.DataFrame()
for column in x_columns:
    df2[column]=df[column]

df2.head(5)
```

```python
X = df2
y = df['Stage']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(
    X,y,test_size=0.33,random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train,y_train)
```

```python
y_pred = log_reg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
